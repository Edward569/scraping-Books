```python
import requests 
from bs4 import BeautifulSoup as bs4
import pandas as pd

pages=[]
prices=[]
stars=[]
titles=[]
urles=[]

pages_to_scrape=1

for i in range(1,pages_to_scrape+1):
    url=('http://books.toscrape.com/catalogue/category/books/science_22/index.html').format(i)
    pages.append(url)
    print(pages)


#وده بستخدمه لو عاوز اسكراب كذا صفحة مع بعض بس لو هسكراب صفحة واحدة مش لازم ياعني
```

```python
for item in pages:
    page=requests.get(item)
    soup=bs4(page.text,'html.parser')
    print(soup.prettify())
```

```python
for i in soup.findAll('h3'):
    ttl=i.getText()
    titles.append(ttl)

for i in soup.findAll('p',class_='price_color'):
    prc=i.getText()
    prices.append(prc)

for i in soup.findAll('p',class_='star-rating'):
    for k,v in i.attrs.items():
        star=v[1]
        stars.append(star)
        
        
        

#الكلاس اللي جواه كذا كلاس بيتعامل معاملة الديكشناري و بيتكتب كدا 
```

```python
data={'titles':titles,'prices':prices,'stars':stars}
print(data)
```

```python
df=pd.DataFrame(data = data)
df
```
